```python
import numpy as np
```

1. Textfile

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
np.savetxt('data.txt', arr)
```

2. Numpy file

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
np.save('data.npy', arr)
```

3. CSV

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
np.savetxt('data.csv', arr, delimiter = ',')
```

4. json

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
np.savetxt('data.json', arr, delimiter = ',', header = '{"data" : [ ', footer = ']}')
```

5. yaml

```python
arr = np.array([[1, 2, 3],[4, 5, 6]])
np.savetxt('data.yaml', arr, delimiter = ',', header = '- data : ')
```
