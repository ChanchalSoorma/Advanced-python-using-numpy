```python
import numpy as np
```

1. textfile

```python
data = np.loadtxt('data.txt')
print(data)
```

2. npy

```python
data = np.load('data.npy')
print(data)
```

3. csv

```python
data = np.loadtxt('data.csv')
print(data)
```

4. json

```python
data = np.loadtxt('data.json', delimiter = ',')
print(data)
```

5. yaml

```python
data = np.loadtxt('data.yaml', delimiter = ',')
print(data)
```
