```python
import numpy as np
```

1. Flattening

```python
arr = np.array([[1, 2, 3],[ 4, 5, 6]])
print(arr)
print('-' * 10)

flat_arr = arr.flatten()
print(flat_arr)
```

2. Reshaping

```python
arr = np.array([[1, 2, 3],[ 4, 5, 6]])
print(arr)
print('-' * 10)

reshaped_arr = arr.reshape(3,2)
print(reshaped_arr)
```

3. Splitting

- vsplit

```python
arr = np.array([[1, 2, 3],[ 4, 5, 6],[3, 5, 7],[1, 9, 8]])
print(arr)
print('-' * 10)

vs_split = np.vsplit(arr, 4)
print(vs_split)
print('-' * 10)

for i in np.vsplit(arr,4):
  print(i)
```

- hsplit

```python
arr = np.array([[1, 6, 2, 3],[ 4, 1, 5, 6],[3, 2, 5, 7],[1, 4, 9, 8]])
print(arr)
print('-' * 10)

hs_split = np.hsplit(arr, 4)
print(hs_split)
print('-' * 10)

for i in np.hsplit(arr,2):
  print(i)
```

- Split

```python
arr = np.array([[1, 6, 2, 3],[ 4, 1, 5, 6],[3, 2, 5, 7],[1, 4, 9, 8]])
print(arr)
print('-' * 10)

split = np.split(arr, 4 , axis = 0)       # prints horizontal value
print(split)
print('-' * 10)

for i in np.split(arr,4 , axis = 1):      # prints vertical value
  print(i)
```

4. Merging

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 2, 7, 4])
arr3 = np.array([1, 6, 1, 7])
arr4 = np.array([4, 5, 4, 9])

concat_arr = np.concatenate((arr1, arr2, arr3, arr4))
print(concat_arr)
```

- hstack

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 2, 7, 4])
arr3 = np.array([1, 6, 1, 7])
arr4 = np.array([4, 5, 4, 9])

h_stack = np.hstack((arr1, arr2, arr3, arr4))
print(h_stack)
```

- vstack

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 2, 7, 4])
arr3 = np.array([1, 6, 1, 7])
arr4 = np.array([4, 5, 4, 9])

v_stack = np.vstack((arr1, arr2, arr3, arr4))
print(v_stack)
```

- stack

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 2, 7, 4])
arr3 = np.array([1, 6, 1, 7])
arr4 = np.array([4, 5, 4, 9])

_stack = np.stack((arr1, arr2, arr3, arr4), axis = 0)
print(_stack)
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 2, 7, 4])
arr3 = np.array([1, 6, 1, 7])
arr4 = np.array([4, 5, 4, 9])

_stack = np.stack((arr1, arr2, arr3, arr4), axis = 1)
print(_stack)
```
