1. None to np.nan

```python
import numpy as np
arr = np.array([1,2,3,None, 4,5,None])
print(arr)

arr[arr == None] = np.nan
print(arr)
```

2. Checking for Null Values

```python
arr = np.array([1,2,3,np.nan, 4,5,np.nan])
print(arr)

print(np.isnan(arr))
```

3. Removing Null Values

```python
arr = np.array([1,2,3,np.nan, 4,5,np.nan])

print(np.isnan(arr))
print(~np.isnan(arr))

arr = arr[~np.isnan(arr)]
print(arr)
```

4. Filling Null Values with Fixed Values

```python
arr = np.array([1,2,3, np.nan, 4,5,np.nan])
print(arr)

arr[np.isnan(arr)] = 0
print(arr)
```

5. Filling Null Values with Average

```python
arr = np.array([21,22,22, np.nan, 24,25,np.nan])
print(arr)

arr[np.isnan(arr)] = int(np.nanmean(arr))
print(arr)
```
