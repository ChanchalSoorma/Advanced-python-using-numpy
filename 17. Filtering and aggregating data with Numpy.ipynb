```python
import numpy as np
```

```python
arr = np.random.randint(1,11, size = (5, 5))
arr
```

1. Filtering

- Filtering elements greater than 5

```python
arr[arr>5]
```

- Filtering the elements that are even

```python
arr%2
```

```python
arr%2 == 0
```

- Filter elements that are greater than 5 and are even

```python
arr[(arr>5) & (arr%2 == 0)]
```

- Filter elements that are divisible by 3 or 7

```python
arr[(arr%3 == 0) | (arr%7 == 0)]
```

- Filtering Based on Conditions of Another Array

```python
arr = np.random.randint(1,11,size=(5,5))
arr1 = np.array([True, False, True,False,True])
print(arr[arr1])
```

- Filtering Based on Specific Row Conditions

```python
arr = np.random.randint(1,11,size=(5,5))
filtered_arr = arr[2][arr[2,:] > 5]
print(filtered_arr)
```

2. Aggregating data

- Summation

```python
arr = np.array([1,2,3,4,5])
print(np.sum(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.sum(arr))
print(np.sum(arr, axis = 0))
print(np.sum(arr, axis = 1))
```

- Mean Calculation

```python
arr = np.array([1,2,3,4,5])
print(np.mean(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.mean(arr))
print(np.mean(arr, axis = 0))
print(np.mean(arr, axis = 1))
```

- Standard Deviation

```python
arr = np.array([1,2,3,4,5])
print(np.std(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.std(arr))
print(np.std(arr, axis = 0))
print(np.std(arr, axis = 1))
```

- Minimum and Maximum

```python
arr = np.array([1,2,3,4,5])
print(np.min(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.min(arr))
print(np.min(arr, axis = 0))
print(np.min(arr, axis = 1))

print('-'*20)

arr = np.array([1,2,3,4,5])
print(np.max(arr))

arr = np.array([[1,2,3],[4,5,6]])
print(np.max(arr))
print(np.max(arr, axis = 0))
print(np.max(arr, axis = 1))
```
