Descriptive Statistics Computation:

```python
import numpy as np

# Generating a random array of 100 integers between 1 and 100
arr = np.random.randint(low=1, high=101, size=100)

# Computing mean and median
mean = np.mean(arr)
median = np.median(arr)

# Computing variance and standard deviation
variance = np.var(arr)
std_dev = np.std(arr)

# Computing quartiles (Q1, Q2, Q3) and interquartile range (IQR)
q1 = np.percentile(arr, 25)
q2 = np.percentile(arr, 50)  # Median is also the 50th percentile
q3 = np.percentile(arr, 75)
iqr = q3 - q1

# Computing skewness and kurtosis
skewness = np.mean((arr - mean) ** 3) / (std_dev ** 3)
kurtosis = np.mean((arr - mean) ** 4) / (std_dev ** 4)

# Printing the computed statistics
print('Mean      : ', mean)
print('Median    : ', median)
# print('Mode      : ', mode)  # Mode calculation is commented out as it's not directly available in NumPy
print('-'*20)

print('Variance  : ', variance)
print('Standard  : ', std_dev)
print('-'*20)

print('Q1        : ', q1)
print('Q2        : ', q2)
print('Q3        : ', q3)
print('-'*20)

print('IQR       : ', iqr)
print('Skewness  : ', skewness)
print('Kurtosis  : ', kurtosis)
```
