```python
import numpy as np
```

1. Generating Random Numbers:

```python
print(np.random.rand())
print('-' * 10)

print(np.random.rand(3))
print('-' * 10)

print(np.random.randint(10))
print('-' * 10)

print(np.random.randint(10, size=(3, 5)))
```

2. Random numbers with Seed

```python
np.random.seed(2)

print(np.random.rand())
print('-' * 10)

print(np.random.rand(3))
print('-' * 10)

print(np.random.randint(10))
print('-' * 10)

print(np.random.randint(10, size=(3, 5)))
print('-' * 10)
```

```python
np.random.seed(5)

print(np.random.rand())
print('-' * 10)

print(np.random.rand(3))
print('-' * 10)

print(np.random.randint(10))
print('-' * 10)

print(np.random.randint(10, size=(3, 5)))
print('-' * 10)
```

3. Seed in Loops:

```python
np.random.seed(5)

# Run the loop 10 times
for i in range(10):

    # Generate an array of 1000 random integers (either 0 or 1)
    arr = np.random.randint(2, size=1000)

    # Count and print the number of ones and zeros in the array
    print(len(arr[arr == 1]), len(arr[arr == 0]))
```
