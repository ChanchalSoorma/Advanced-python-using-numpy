```python
import numpy as np
```

1. Range (for list)

```python
r = range(10)
print(r)

r = range(0,10, 2)
print(r)

r = range(10,0, -2)
print(r)

print(6 in range(10))
print(list(range(10)))
```

2. Arange (for array)

```python
arr = np.arange(0,10,2)
print(arr)

arr1 = np.arange(10,0,-1)
print(arr1)

arr2 = np.arange(0,10,.5)
print(arr2)
```

3. Linspace (evenly spaced values between a start and stop value.)

```python
l = np.linspace(0,10,5)
print(l)

l = np.linspace(1,10,10)
print(l)

l = np.linspace(1,10,10, endpoint = False)
print(l)

l,s= np.linspace(1,5,10, retstep = True)
print(l)
print(s)
```

4. Logspace (evenly spaced values on a logarithmic scale.)

```python
arr = np.logspace(0,10,5)
print(arr)

arr = np.logspace(0,10,5, endpoint = False)
print(arr)

arr = np.logspace(0,10,5, base = 2)
print(arr)
```
