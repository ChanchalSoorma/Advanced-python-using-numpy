```python
import numpy as np
```

1. CONVERTING DATA STRUCTURES TO NUMPY ARRAYS

1.1 Converting a tuple to an array

```python
tpl = (1, 2, 3)
print(tpl)
print(type(tpl))
print('----------')

arr = np.array(tpl)
print(arr)
print(type(arr))
print('----------')
```

1.2 Converting a list into an array

```python
lst = [1, 2, 3]
print(lst)
print(type(lst))
print('----------')

arr = np.array(lst)
print(arr)
print(type(arr))
print('----------')
```

1.3 Converting a dictionary into an array

```python
dict = {1 : 'Apple', 2 : 'Banana', 3 : 'Cherry'}
print(dict)
print(type(dict))
print('----------')

arr = np.array(dict)
print(arr)
print(type(arr))
print('----------')

arr_k = np.array(list(dict.keys()))
arr_v = np.array(list(dict.values()))
print(arr_k)
print(arr_v)
```

1.4 Converting a set into an array

```python
set_ = {1, 2, 3}
print(set_)
print(type(set_))
print('----------')

arr = np.array(list(set_))
print(arr)
print(type(arr))
```

```python
arr = np.array('Hello World')
arr1 = np.array(list('Hello World'))

print(arr)
print(type(arr))
print('-' * 10)

print(arr1)
print(type(arr1))
```

2. CONVERTING THE DATATYPES OF NUMPY ARRAYS

2.1 Create an array of integers

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr)

arr_ = arr.astype('float')
print(arr_)

arr_ = arr.astype('str')
print(arr_)

arr_ = arr.astype('bool')
print(arr_)
```

2.2 Create an array of floats

```python
arr = np.array([1. , 2. , 3. , 4. , 5.])
print(arr)

arr_ = arr.astype('int')
print(arr_)

arr_ = arr.astype('str')
print(arr_)

arr_ = arr.astype('bool')
print(arr_)
```

2.3 Create an array of strings

```python
arr = np.array(['1', '2', '3', '4', '5'])
print(arr)

arr_ = arr.astype('float')
print(arr_)

arr_ = arr.astype('int')
print(arr_)

arr_ = arr.astype('bool')
print(arr_)
```

2.4 Create an array of booleans

```python
arr = np.array([True, False, False, False, True])
print(arr)

arr_ = arr.astype('float')
print(arr_)

arr_ = arr.astype('str')
print(arr_)

arr_ = arr.astype('int')
print(arr_)
```
